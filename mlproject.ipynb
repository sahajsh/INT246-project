```python
#!/usr/bin/env python
# coding: utf-8

# In[34]:


import pandas as pd
import numpy as np
df=np.random.randint(30,100,(10,3))#creating random dataset
df=pd.DataFrame(df,columns=['no_of_vehicle','avg_speed','avg_distance_btw_them'])
df.to_csv('data.csv')
df=df.values
prob_dim=(len(df))
pop_size=30
data_set=[[0 for x in range(prob_dim)] for x in range(pop_size)]

def generate(df):
    no_of_lane=prob_dim
    for i in range(pop_size):
        for j in range(no_of_lane):
            data_set[i][j]=((df[j][0]*df[j][1])//df[j][2])*np.random.randint(1,7)#(between 0-8*(timeconstant_for_lane))
            # print(data_set[i][j])
    return data_set


pop=generate(df)


# In[35]:


"""f(x)=x1^2+x2^2+x3^2
0=<x<=8
size=6"""

pop=np.array(data_set)
"""initialization for random samples
  x=xmin+rand(0,1)*(xmax-xmin)"""
for i in pop:
    print(i)


# In[36]:


def objective(pop):
    f=[]
    for i in range(pop.shape[0]):
        f.append(pop[i,0]**2+pop[i,1]**2+pop[i,2]**2)
    return f
def fitness(pop):
    fit=[]
    f=objective(pop)
    for i in f:
        if(i<0):
            fit.append(1-i)
        else:
            fit.append(1/(1+i))
    return fit
trial=[0]*pop.shape[0]
best=pop[np.argmax(fitness(pop)),:]
print("Population is ",pop)
print("Fitness of population ",fitness(pop))
print("Trial value of population ",trial)
print("best soluntion is ",best)


# In[37]:


#Employee Bee phase
import random
lb=0
up=8
def employee(pop,lb,ub,fit,trial):
    for i1 in range(pop.shape[0]):
        j=random.randint(0,2)
        p=random.randint(0,2)
        while(i1==p):
            p=random.randint(0,2)
        new=pop[i1,j]+np.random.uniform(low=-1,high=1,size=(1,))*(pop[i1,j]-pop[p,j])
        if(new>ub):
            new=ub
        if(new<lb):
            new=lb
        new_pop=np.array([pop[i1,:].copy()])
        new_pop[0,j]=new
        #print(new_pop.shape)
        f=fitness(new_pop)
        if(f>fit[i1]):
            pop[i1,:]=new_pop
            trial[i1]=0
        else:
            trial[i1]=trial[i1]+1
    return pop

print(employee(pop,lb,up,fitness(pop),trial))
print(trial)


# In[ ]:





# In[ ]:





# In[ ]:
```
